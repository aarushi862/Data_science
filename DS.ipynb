```python
# hypnothesis testing for this we import scipy.stats
# a teacher claims that the mean score of student in his class is greater than 82 with a std of 20.if a sample of 81 student are selected with a mean score of 90.
# z test
x=90
u=82
std=20
n=81
alpha=0.05

```

```python
z_cal=(x-u)/(std/np.sqrt(n))
z_cal
```

```python
z_table=st.norm.ppf(1-0.05)
z_table
```

```python
if z_cal>z_table:
  print("reject null hypothesis or ha is right")
else:
  print("accept null hypothesis or ho is right")
```
